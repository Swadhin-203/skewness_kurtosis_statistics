# Skewness 

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
def normalize(column):
    return (column-column.min())/(column.max()-column.min())
```

```python
df["normalize_age"]=normalize(df["age"])
```

```python
df["normalize_age"]
```

```python
df["left_skewed"]=df["normalize_age"]*0.5
```

```python
df["right_skewed"]=df["normalize_age"]*2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["normalize_age","left_skewed","right_skewed"]])
plt.ylim(0,1)
plt.xlabel("column")
plt.ylabel("value")
```

```python
# Example
df["left_skewed"]=df["normalize_age"]-0.2
df.loc[df["left_skewed"]<-0.2,"left_skewed"]-=0.5
df.loc[df["right_skewed"]>1.2,"right_skewed"]+=0.5
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["normalize_age","left_skewed","right_skewed"]])
plt.ylim(-0.5,1.5)
plt.xlabel("column")
plt.ylabel("value")
```

```python
df=sns.load_dataset("iris")
```

```python
df["normalize_sepal_width"]=normalize(df["sepal_width"])
```

```python
left_skewed_data=np.random.normal(loc=-0.2,scale=0.05,size=len(df))
```

```python
df["left_skewed"]=df["normalize_sepal_width"]-left_skewed_data
```

```python
right_skewed_data=np.random.normal(loc=-0.2,scale=0.05,size=len(df))
df["right_skewed"]=df["normalize_sepal_width"]+right_skewed_data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["normalize_sepal_width","left_skewed","right_skewed"]])
plt.ylim(-0.5,1.5)
plt.xlabel("column")
plt.ylabel("value")
```

# Kutorist
## It gives of shape of the data
## It gives the peakness of the data
## It is of threee types
## 1-Mesokurtic:-Normal Distiburtion Data
## 2-Leptokurtic(High Kurtosis):-Here the kurtosis value is always greater than 3 . It has a heavy tail and a higher peak
## 3-Platykurtic(Low Kurtosis):-Kurtosis is always less than 3. It has a lighter or thiner tail

```python
df=sns.load_dataset("iris")
def normalize(column):
    return (column-column.min())/(column.max()-column.min())
iris_normalized=df["sepal_length"]
normalize_data=normalize(df["sepal_length"])
plt.hist(normalize_data,bins=20,edgecolor="black")
plt.xlabel("Normalize Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Normalized Sepal Length")
```

```python
high_kurotosis_data=(df["sepal_length"]-df["sepal_length"].mean())*5
plt.hist(normalize_data,bins=20,edgecolor="black")
plt.xlabel("High Kurtosis Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of high Kurtosis Sepal Length")
```
